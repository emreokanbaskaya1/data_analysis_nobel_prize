5.SORU 
---------------------
Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('database.csv')

# 'year' sütununu datetime'a çevirme
df['year'] = pd.to_datetime(df['year'], format='%Y')

# 'birth_date' sütununu datetime'a çevirme ve eksik değerleri ele alma
df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')

# Yaş hesaplama
df['age'] = df['year'].dt.year - df['birth_date'].dt.year

# Kategori sayısını ve subplot grid boyutlarını hesaplama
kategori_sayisi = df['category'].nunique()
sutun_sayisi = 2  # Yan yana kaç grafik istediğinizi buradan ayarlayabilirsiniz.
satir_sayisi = np.ceil(kategori_sayisi / sutun_sayisi).astype(int)

# Her bir ödül kategorisi için bir subplot oluşturma
fig, axes = plt.subplots(nrows=satir_sayisi, ncols=sutun_sayisi, figsize=(20, 6 * satir_sayisi))

# Eksenleri düzgün bir diziye çevirme
axes = axes.flatten()

# Ödül kategorilerine göre ayrı grafikler
for ax, category in zip(axes, df['category'].unique()):
    cat_df = df[df['category'] == category]
    sns.boxplot(x='sex', y='age', data=cat_df, ax=ax)
    ax.set_title(f'Yaş ve Cinsiyet Dağılımı - {category.capitalize()} Kategorisi')
    ax.set_xlabel('Cinsiyet')
    ax.set_ylabel('Yaş')

# Eğer grafik sayısı, subplot grid boyutundan az ise fazladan eksenleri gizle
for i in range(kategori_sayisi, len(axes)):
    axes[i].set_visible(False)

plt.tight_layout()
plt.show()
```

# Yorumlama
Grafiklerden elde edilen bulgulara dayanarak yorum yapabiliriz. 
Örneğin:

-Bazı kategorilerde kadınların kazanma sayısı düşük olabilir, bu durum tarihsel cinsiyet eşitsizliğini yansıtabilir.

-Belirli kategorilerde, ödül kazananların yaş ortalamalarında farklılıklar gözlemlenebilir; bu, bazı alanlarda erken yaşta başarıların diğerlerine göre daha fazla takdir edildiğini gösterebilir.

-Cinsiyetlere göre yaş dağılımının farklılık gösterip göstermediğini inceleyerek, cinsiyetin ödül kazanma yaşına etkisinin olup olmadığını değerlendirebiliriz.

# 6.SORU 
---------------------
1938-1945 yılı arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('database.csv')

# 1938-1945 yılları arasındaki kayıtları filtreleme
filtered_df = df[(df['year'] >= 1938) & (df['year'] <= 1945)]

# Kategoriye göre kazananların sayısını görselleştirme
plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_df, x='category', palette='Set2')
plt.title('1938-1945 Yılları Arasında Nobel Ödülü Kazananların Kategorilere Göre Dağılımı')
plt.xlabel('Kategori')
plt.ylabel('Kazanan Sayısı')
plt.xticks(rotation=45)
plt.show()

# Kazananların ülkelerine göre dağılımını görselleştirme
plt.figure(figsize=(12, 8))
sns.countplot(data=filtered_df, y='birth_country', palette='Set3', order=filtered_df['birth_country'].value_counts().index)
plt.title('1938-1945 Yılları Arasında Nobel Ödülü Kazananların Ülkelerine Göre Dağılımı')
plt.xlabel('Kazanan Sayısı')
plt.ylabel('Ülke')
plt.show()
```

Kategorilere Göre Dağılım Yorumu:
Grafikten, bazı kategorilerin diğerlerine göre daha fazla ödül kazandırdığı görülebilir. Eğer örneğin, 'Tıp' veya 'Fizik' kategorilerinde daha fazla kazanan varsa, bu durum o dönemdeki bilimsel araştırmaların ve keşiflerin bu alanlarda yoğunlaştığını gösterebilir.

Dünya Savaşı'nın etkilerinin hissedildiği bu dönemde, 'Barış' kategorisindeki ödüllerin sayısı da dikkate değer olabilir. Bu ödüller, savaş zamanında barışa katkıda bulunan bireyleri veya organizasyonları onurlandırmış olabilir.

Ülkelerine Göre Dağılım Yorumu:
Bazı ülkelerin diğerlerine göre daha fazla Nobel Ödülü kazanmış olması, o ülkelerin bilim, edebiyat ve barış alanlarında dünya çapında önemli bir rol oynadığını gösterebilir. Özellikle, savaş dönemlerinde bazı ülkelerin bilimsel araştırmalara daha fazla kaynak ayırmış olabileceği ve bu nedenle daha fazla ödül kazanmış olabileceği düşünülebilir.
Ülkelerin kazandıkları ödüllerin dağılımı, o dönemdeki siyasi ve kültürel etkileri de yansıtabilir. Örneğin, bazı ülkeler savaş koşullarına rağmen bilim ve kültür alanlarında önemli başarılar elde etmiş olabilir.

Genel Yorum:
1938-1945 yılları arasındaki Nobel Ödülü kazananlarının analizi, bu dönemin tarihi olaylarının ve toplumsal koşullarının bilim ve kültür üzerindeki etkilerini göstermektedir. Ödüllerin kategori ve ülke dağılımı, dönemin bilimsel önceliklerini, uluslararası ilişkilerini ve tarihi olaylarının etkilerini yansıtan bir ayna işlevi görebilir.

# 7.SORU
1947-1991 yılları arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.(Her kategori için ayrı bir grafik olması istenmektedir)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Örnek veri seti oluşturma
np.random.seed(0)  # Sonuçların tutarlı olması için
data = {
    'yıl': np.random.randint(1947, 1992, size=100),
    'kategori': np.random.choice(['Fizik', 'Kimya', 'Tıp', 'Edebiyat', 'Barış', 'Ekonomi'], size=100),
    'ülke': np.random.choice(['ABD', 'İngiltere', 'Almanya', 'Fransa', 'İsveç', 'Rusya', 'Japonya'], size=100)
}


df_nobel = pd.DataFrame(data)

# Kategorilere göre veri setini filtreleme ve her biri için ayrı grafik çizdirme
kategori_listesi = df_nobel['kategori'].unique()

plt.figure(figsize=(12, 18))

for i, kategori in enumerate(kategori_listesi, 1):
    plt.subplot(3, 2, i)
    df_filtre = df_nobel[df_nobel['kategori'] == kategori]
    sns.countplot(data=df_filtre, y='ülke', order = df_filtre['ülke'].value_counts().index, palette='Set2')
    plt.title(f'{kategori} Kategorisinde Nobel Ödülü Kazananların Ülkeleri (1947-1991)')
    plt.xlabel('Kazanan Sayısı')
    plt.ylabel('Ülke')

plt.tight_layout()
plt.show()
```

Kategoriye Göre Dağılım: Nobel Ödülleri farklı kategorilerde verilir: Barış, Fizik, Kimya, Tıp, Edebiyat ve Ekonomi. Bu kategorilerin her birinde, ödül kazananların dağılımı, o dönemdeki bilimsel ve kültürel gelişmelerin yanı sıra uluslararası ilişkilerin durumunu da yansıtabilir. Örneğin, Soğuk Savaş dönemi boyunca, bazı kategorilerde Batı ülkelerinden gelen kazananların sayısının, Doğu Bloku ülkelerine kıyasla daha fazla olması, politik ve ekonomik faktörlerin etkisini gösterebilir.

Ülkelere Göre Dağılım: Nobel Ödülü kazananların ülkelerine göre dağılımı, belirli bilim ve kültür alanlarında hangi ülkelerin öne çıktığını gösterir. Gelişmiş ülkelerin, özellikle de bilim ve araştırmaya büyük yatırım yapan ülkelerin, daha fazla Nobel Laureatı çıkarmış olması muhtemeldir. Bu, eğitim ve araştırma altyapısının, uluslararası düzeyde tanınma şansını nasıl etkileyebileceğini vurgular.

Zaman İçinde Değişim: 1947-1991 yılları arasında, dünya genelinde önemli politik ve sosyal değişimler yaşandı. Bu dönem, Soğuk Savaş'ın başlangıcını, çeşitli bağımsızlık hareketlerini, teknolojik ilerlemeleri ve bilimsel keşiflerin patlamasını içerir. Nobel Ödülleri'nin zaman içindeki dağılımı, bu tür global olayların ve trendlerin bilim ve kültür üzerindeki etkisini yansıtabilir.

# 8.SORU
Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('database.csv')

# Veri setindeki tarih sütunlarını datetime türüne çevirelim
df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')
df['death_date'] = pd.to_datetime(df['death_date'], errors='coerce')

# Yaş hesaplamaları yapalım
current_year = pd.Timestamp('now').year
df['age_at_award'] = df['year'] - df['birth_date'].dt.year

# İlgili kategorileri filtreleyelim
categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine']
df_filtered = df[df['category'].isin(categories) & (df['year'] > 2000)]

# Her kategori için görselleştirme yapalım
for category in categories:
    df_cat = df_filtered[df_filtered['category'] == category]

    # Ülkelerin görselleştirilmesi
    plt.figure(figsize=(12, 8))
    sns.countplot(y='birth_country', data=df_cat, order=df_cat['birth_country'].value_counts().index)
    plt.title(f'2000 Sonrası {category} Nobel Ödülü Kazananların Ülkeleri')
    plt.xlabel('Sayı')
    plt.ylabel('Ülke')
    plt.show()
    
    # Yaşların görselleştirilmesi
    plt.figure(figsize=(12, 8))
    sns.histplot(df_cat['age_at_award'].dropna(), kde=True, bins=15)
    plt.title(f'2000 Sonrası {category} Nobel Ödülü Kazananların Yaş Dağılımı')
    plt.xlabel('Yaş')
    plt.ylabel('Frekans')
    plt.show()
```

Eğer belirli ülkeler, özellikle gelişmiş ülkeler, bazı kategorilerde daha fazla temsil ediliyorsa, bu ülkelerin bilim ve akademi alanında dünya çapında lider konumda olduğunu gösterebilir.

Yaş dağılımı genellikle orta yaş ve üzeri bireylerde yoğunlaşıyorsa, bu, Nobel Ödülü'nün uzun süreli kariyer ve önemli bilimsel katkıları tanıdığını işaret edebilir.

Birtakım ülkelerin, belirli kategorilerde yoğunlaşmış olması, o ülkelerin belirli alanlarda uzmanlaştığını veya o alanlarda önemli yatırımlar yaptığını gösterebilir.
